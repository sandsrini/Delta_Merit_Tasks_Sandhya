```python
driver=webdriver.Chrome(executable_path=driverPath)
driver.get("https://ai.fmcsa.dot.gov/hhg/Search.asp?ads=a&lan=en")
driver.maximize_window()
driver.find_element_by_xpath("//label[contains(text(),'State :')]//following::select[contains(@class,'search_select')]").click()
writer=(pd.ExcelWriter('States.xlsx',engine='xlsxwriter'))
for i in range(1,97):
    lst=list()
    driver.find_element_by_xpath("//option[contains(text(),'Please select state')]//following::option[{}]".format(i)).click()
    driver.find_element_by_xpath("//label[contains(text(),'State :')]//following::input[contains(@name,'Submit')]").click()
    #time.sleep(1)
    state=driver.find_elements_by_xpath("//tr[contains(@scope,'row')]")
    for j in state:
        lst.append(j.text.split("  "))
    df=pd.DataFrame(lst,columns=['Company_Name','Headquarters','Company_Type','Fleet_Size'])
    df.to_excel(writer, sheet_name=str(i), index=False)
    driver.back()
writer.save()
    
driver.close()
#print(lst1)
#len(lst1[0])
```
