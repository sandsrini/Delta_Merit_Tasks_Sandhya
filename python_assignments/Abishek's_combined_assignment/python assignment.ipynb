### Task 1 Collecting Batsmen details in a excel file with 3 sheets

```python
driver=webdriver.Chrome(executable_path=driverPath)
driver.get("https://www.cricbuzz.com/")
driver.maximize_window()
driver.find_element_by_xpath("//header/div[1]/nav[1]/div[5]/a[1]").click()
a=driver.find_elements_by_xpath("//header/div[1]/nav[1]/div[5]/a[1]//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
lst=list()
for i in a:
    lst.append(i.text)
    print(i.text)
driver.find_element_by_xpath("//a[@id='batsmen-odis-tab']").click()
b=driver.find_elements_by_xpath("//a[@id='batsmen-tests-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
lst2=list()
for i in b:
    lst2.append(i.text)
    print(i.text)
driver.find_element_by_xpath("//a[@id='batsmen-t20s-tab']").click()
c=driver.find_elements_by_xpath("//a[@id='batsmen-t20s-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
lst3=list()
for i in c:
    lst3.append(i.text)
    print(i.text)
print(lst3)
```

```python
lst_split=list()
for i in lst:
    lst_split.append(i.split("\n"))
print(lst_split)
lst2_split=list()
for i in lst2:
    lst2_split.append(i.split("\n"))
print(lst2_split)
lst3_split=list()
for i in lst3:
    lst3_split.append(i.split("\n"))
print(lst3_split)
    
```

```python
df=pd.DataFrame(lst_split,columns=['Position','NAN','Player','Country','Rating'])
df1=pd.DataFrame(lst2_split,columns=['Position','NAN','Player','Country','Rating'])
df2=pd.DataFrame(lst3_split,columns=['Position','NAN','Player','Country','Rating'])
```

```python
df.dropna(axis='rows',thresh=4,inplace=True)
df1.dropna(axis='rows',thresh=4,inplace=True)
df2.dropna(axis='rows',thresh=4,inplace=True)
```

```python
writer =pd.ExcelWriter('test.xlsx',engine='xlsxwriter')
```

```python
df.to_excel(writer,sheet_name='Batsman_Test', index=False)
df1.to_excel(writer,sheet_name='Batsman_Odi', index=False)
df2.to_excel(writer,sheet_name='Batsman_T20', index=False)
```

```python
writer.save()
```
